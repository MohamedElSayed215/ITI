```python
#Problem(2):Distinct Numbers
filtered_list=[]
def distinct_numbers(list):
    for number in list :
        if number not in  filtered_list :
            filtered_list.append(number)
    filtered_list.sort()
    return filtered_list

print (distinct_numbers([3, 2, 2, 1, 0, 2, 4, 6, 9, 3, 3, 1] ))


```

```python
#Problem(3):The length of the longest in the list 
def length_of_longest(list):
    
    lengths =[len(element) for element in list ]
    return max(lengths)


my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"] 
#my_list = ["first", "second", "fourth", "eleventh"] 
print(f' the result is :{length_of_longest(my_list)}')
```

```python
#Problem(4):All the longest in the list 
def all_the_longest(list):
    longest=[]
    lengths =[len(element) for element in list ]
    indices = [list.index(value) for value in (list) if len(value) == max(lengths)]
    #print(indices)
    for i in range(len(indices)):
        longest.append(list[indices[i]])
    print(longest)    
             

            
my_list = ["first", "second", "fourth", "eleventh"] 
#my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"] 
all_the_longest(my_list)
```
