```python
#(1)

def triangle(num):
    for i in range(1,num+1):
        print(i*'#')

triangle(6)
```

```python
#(2)
def max_num(num1,num2,num3):
    if num1 >= num2 and num1 >= num3 :
        return num1 
    elif num2 > num1 and num2 >num3 :
        return num2
    else :
        return num3
print(max_num(10,10,5))
```

```python
#(2) Another Way ! 
def max_num(num1,num2,num3):
    if(num1>num2):
        if(num1>num3):
            return num1 
    elif(num2>num1):
        if(num2>num3):
            return num2
        else:
            return num3
    

print(max_num(4,9,3))     
        
```

```python
#(3)
def sum_series(num):
    sum=0
    for i in range(num+1):
        if (i%2==0):
            sum+=-1*i
        else:
            sum+=i

    return sum

print(sum_series(5))
```

```python
#(4)
def fact_num(num):
    if (num > 0):
        fact=1
        for i in range (1,num+1):
            fact*=i
        print(f'result = {fact}')
    else : 
        print('Invalid value ! ')         


fact_num(5)
```

```python
#(5)

def IsPalindrome(word):
    word=word.lower()
    reversed_word = ""
    for char in word:
        reversed_word =  char + reversed_word 
        print(reversed_word)        
    if(word == reversed_word):
        print("Yes , it is Palindrome ! ")
    else :
        print("Yes , it isn't Palindrome ! ")   

IsPalindrome('Mohamed ElSayed')
IsPalindrome('Madam')
    

```
