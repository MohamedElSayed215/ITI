```python
#### **Problem-Solving: Real-World Scenarios**
#Problem(1):**Student Grades:**
students_data={
    'mohamed' :[10,20,30] , 
    'mostafa' :[40,50,60] ,
    'mahmoud' :[70,80,90]
}
averages=[]
names=[]
def calc_averages(dic):
    for name , values in students_data.items():
        sum=0
        for i in range(len(values)):
            sum+=values[i]
        names.append(name)    
        averages.append(sum//len(students_data))

def hightest_average(dic):
     for average in averages :
         if max(averages) == average :
            print(f'the highest is : {names[averages.index(average)]}')  


calc_averages(students_data)
hightest_average(students_data)

 

    
```

```python
## Problem(2):Shoppint Cart 
import random
shopping_cart={
    'Mobile' :10,
    'TV' : 20,
    'Laptop':30 
}
def add_items(dic):
    name= input('Enter the Item :')
    price= random.randint(0, 100)
    shopping_cart[name]=price
    print(dic)
def cal_total_price(dic):
    cost = 0
    for prices in dic.values():
        cost += prices
    return cost

add_items(shopping_cart)
print(f'the total price is : {cal_total_price(shopping_cart)}')
```

```python
## Problem(3):Employee Database ## 
empolyee_database=[
    {'name': 'mohamed' , 'age' : 24 , 'department':'ece'},
    {'name': 'mostafa' , 'age' : 25 , 'department':'ece'},
    {'name': 'mahmoud' , 'age' : 22 , 'department':'cse'},
    {'name': 'tarek' , 'age' : 26 , 'department':'cse'},
    {'name': 'ahmed' , 'age' : 30 , 'department':'mech'},
    {'name': 'ziad' , 'age' : 33 , 'department':'mech'}
]

def print_with_department(dep_input):
    print(f"all employees of '{dep_input}' department are/is :")
    for info in empolyee_database :
        if info['department'] == dep_input :
            print(info)
def print_averages():
    length = len(empolyee_database)
    sum=0
    for ages in empolyee_database:
        sum+= ages['age']
    return sum // length    

dep_input=input('Enter the specific department :')
print_with_department(dep_input)
print(f' the average is {print_averages()}')
```

```python
#### **Challenge Section**
## **Word Frequency Counter:**
dic={}
words=[]
counters=[]
def word_frequency_counter():
    #usr_input=input('Enter the text :')
    usr_input='Hello World Time New World New Time ' # 1 , 2 , 2 ,2
    usr_input=usr_input.split()
    
    print(usr_input)

    for text in usr_input :
        if text not in words :
            counters.append(0)
            words.append(text)
        
        counters[words.index(text)]+=1  

    for i in range (len(words)):
        dic[words[i]]=counters[i]
    print(dic)    
word_frequency_counter()    
```

```python
### Proble(4): Anagram Checker ### 
def anagram_checker(str1,str2):
    sum1=0
    sum2=0
    for i in str1 :
        sum1+=ord(i)
    for i in str2 :
        sum2+=ord(i)
    
    return sum1==sum2
   
print (anagram_checker('listen','silent'))
```
